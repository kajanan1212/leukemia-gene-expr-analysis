```python
import pandas as pd
```

```python
y = pd.read_csv("./data/actual.csv")
X_train_t = pd.read_csv("./data/data_set_ALL_AML_train.csv")
X_test_t = pd.read_csv("./data/data_set_ALL_AML_independent.csv")
```

```python
y["cancer"] = y["cancer"].replace({"ALL": 0, "AML": 1})
```

```python
y_train = y[y["patient"] <= 38]
y_test = y[y["patient"] > 38]
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
X_train_t["X"] = X_train_t["Gene Accession Number"].str.strip(" \n")
X_test_t["X"] = X_test_t["Gene Accession Number"].str.strip(" \n")

X_train_t = X_train_t.drop(columns=["Gene Accession Number", "Gene Description"] + list(X_train_t.filter(like="call").columns))
X_test_t = X_test_t.drop(columns=["Gene Accession Number", "Gene Description"] + list(X_test_t.filter(like="call").columns))
```

```python
X_train_t.shape
```

```python
X_test_t.shape
```

```python
X_train = pd.melt(X_train_t, id_vars=["X"], var_name="patient", value_name="value")
X_train["patient"] = X_train["patient"].astype(int)
X_train = X_train.pivot(index="patient", columns="X", values="value")
X_train = X_train.reset_index()
X_train = X_train.sort_values(by="patient")
X_train.columns.name = None

X_train.shape
```

```python
X_test = pd.melt(X_test_t, id_vars=["X"], var_name="patient", value_name="value")
X_test["patient"] = X_test["patient"].astype(int)
X_test = X_test.pivot(index="patient", columns="X", values="value")
X_test = X_test.reset_index()
X_test = X_test.sort_values(by="patient")
X_test.columns.name = None

X_test.shape
```

```python
train = pd.merge(X_train, y_train, on="patient")
train
```

```python
test = pd.merge(X_test, y_test, on="patient")
test
```

```python
train.to_csv("./data/train.csv", index=False)
test.to_csv("./data/test.csv", index=False)
```
